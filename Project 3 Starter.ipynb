```python
from sqlalchemy import create_engine
import pandas as pd
import os
import re
```

```python
engine = create_engine('postgresql://postgres:warroo@localhost:5432/tennis')
```

```python
for filename in os.listdir('tennis'):
    file_str = os.path.join(filename)
    print(file_str)
    re_names = re.search('^(\S{2,3})\S*(?=Open)?\S+-((?:wo)?men)',file_str).groups()
    name = '_'.join(re_names).lower()
    load_df = pd.read_csv('tennis/'+file_str)
    load_df.to_sql(name,engine,index=False,if_exists='replace')
```

```python
query = """
SELECT * FROM fre_men
"""
df = pd.read_sql(query,engine)
pd.set_option('display.max_rows', 30)
print(df.shape)
df.head(30)
```
